```python
def calculate_mad(data):
    median = np.median(data)
    deviations = np.abs(data - median)
    mad = np.median(deviations)
    return mad
```

```python
def detect_outliers_hampel(data, k=2):

    median = np.median(data)
    mad = calculate_mad(data)
    threshold = k * 1.4826 * mad  # 1.4826 is a scaling factor for MAD

    outliers = np.abs(data - median) > threshold
    return outliers
```

```python
columns_of_interest = data.iloc[:,2:32]
```

```python
for column in columns_of_interest:
    outliers = detect_outliers_hampel(data[column])
    print(f'Column: {column}, Number of Outliers: {np.sum(outliers)}')
```

```python
X = data.iloc[:,2:32].values
Y = data.iloc[:,1].values
```

```python
from sklearn import preprocessing

RobustScaler = preprocessing.RobustScaler()
X_scale = RobustScaler.fit_transform(X)

X_scale
```

```python
X_scale = pd.DataFrame(X_scale,columns=data.iloc[:,2:32].columns)
X_scale
```

```python
for column in columns_of_interest:
    outliers = detect_outliers_hampel(X_scale[column])
    print(f'Column: {column}, Number of Outliers: {np.sum(outliers)}')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(30,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train,
          batch_size=32, epochs=100,
          validation_data=(X_val, Y_val))
```

```python
model.evaluate(X_test, Y_test)[1]
```

```python
Y_pred = model.predict(X_test)
```

```python
import seaborn as sns
from sklearn import metrics

Y_pred = (Y_pred > 0.7)
accuracy = metrics.accuracy_score(Y_test, Y_pred)
f1_score = metrics.f1_score(Y_test, Y_pred)
recall = metrics.recall_score(Y_test, Y_pred)
precision = metrics.precision_score(Y_test, Y_pred)

print("Accuracy:", accuracy)
print("F1 Score:", f1_score)
print("Recall:", recall)
print("Precision:", precision)
```
